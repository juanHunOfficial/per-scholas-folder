# Description:
You have been provided with data regarding your monthly expenses in various categories. Your task is to create a bar chart to visualize the distribution of expenses across different categories.


```python
import matplotlib.pyplot as plt
```

```python
expense_categories = ['Groceries', 'Utilities', 'Transportation', 'Dining Out','Entertainment']
amount_spent = [500, 300, 200, 400, 250]
```

```python
plt.barh(expense_categories, amount_spent, color='skyblue', height=0.4)

plt.xlabel('Expense Categories')
plt.ylabel('Amount Spent')
plt.title('Monthly Expenses Distribution')

plt.grid(True, axis='y')
plt.show()
```
