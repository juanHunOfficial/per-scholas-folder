## Task 7

```python
# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='month_number', y='total_profit', data=df)

# Customize plot
plt.title('Company profit per month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True, linestyle=':', linewidth=0.8, color='gray') 
plt.tight_layout()
plt.show()
```

## Task 8 

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(df['month_number'], df['total_profit'], marker='o', markeredgecolor='black',markerfacecolor='black', linestyle=':', linewidth=3, color='red')

# Customization
plt.title('Company Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.legend(loc='lower right')
plt.grid(True, linestyle=':', linewidth=0.8, color='gray')  
plt.tight_layout()
plt.show()
```

## Task 9

```python
# Plot multiple lines (for different products)
plt.figure(figsize=(10, 6))

# Plot for each product
plt.plot(df['month_number'], df['facecream'], marker='o', linestyle='-', label='Facecream', color='blue')
plt.plot(df['month_number'], df['facewash'], marker='o', linestyle='-', label='Facewash', color='orange')
plt.plot(df['month_number'], df['toothpaste'], marker='o', linestyle='-', label='Toothpaste', color='green')
plt.plot(df['month_number'], df['bathingsoap'], marker='o', linestyle='-', label='Bathing Soap', color='red')
plt.plot(df['month_number'], df['shampoo'], marker='o', linestyle='-', label='Shampoo', color='purple')
plt.plot(df['month_number'], df['moisturizer'], marker='o', linestyle='-', label='Moisturizer', color='brown')

# Customization
plt.title('Product Profit by Month')
plt.xlabel('Month Number')
plt.ylabel('Sales Units in Numbers')
plt.grid(True, linestyle=':', linewidth=0.8, color='gray')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

## Task 10

```python
# Plot
plt.plot(df['month_number'], df['bathingsoap'], marker='o', linestyle=' ', label='Bathing Soap', color='blue')

# Customization
plt.title('Product Profit by Month')
plt.xlabel('Month Number')
plt.ylabel('Sales Units in Numbers')
plt.grid(True, linestyle='--', linewidth=0.8, color='gray')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
